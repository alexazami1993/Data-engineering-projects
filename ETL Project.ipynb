```python
client = bigquery.Client(project='dark-arcade-431219-f6')
target_table = 'sample_database.dept_manager'

job_config = bigquery.LoadJobConfig(
    skip_leading_rows=1,
    source_format=bigquery.SourceFormat.CSV
    #autodetect=True
)

import os
#file vars
os.chdir('/users/hasibazami/Desktop')
cur_path = os.getcwd()
file = 'dept_manager_data.csv'
file_path = os.path.join(cur_path,'etl_data_folder',file)


with open(file_path, 'rb') as source_file:
    load_job= client.load_table_from_file(
    source_file,
    target_table,
    job_config=job_config
    )

load_job.result()

destination_table = client.get_table(target_table)
print(f'you have {destination_table.num_rows} rows in your table')
```
